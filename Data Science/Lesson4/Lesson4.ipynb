##### С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.

```python
t = np.linspace(0, 10, num = 51)
t
```

##### Создайте массив Numpy под названием f, содержащий косинусы элементов массива t

```python
f = np.cos(t)
f
```

##### Постройте линейную диаграмму, используя массив t для координат по горизонтали,а массив f - для координат по вертикали. Линия графика должна быть зеленого цвета.

```python
plt.plot(t,f,'g');
```

##### Выведите название диаграммы - 'График f(t)'. Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'.

```python
t_dict = {'fontsize':24, 'color':'#607D8B'}
l_dict = {'fontsize':8,'fontweight':'bold'}
plt.title('\n График f(t)',fontdict = t_dict, loc = 'left')
plt.xlabel('\n Значения t\n',fontdict = l_dict)
plt.ylabel('\n Значения f\n',fontdict = l_dict)

# Добавим легенду для графика
plt.errorbar(t,f,label='f(t) = cos(t)')

# borderpad - отступ для рамки легенды
# handlelength - длинна линии в легенде
plt.legend(loc = 'best', fontsize = 7.5, borderpad = 1.4, handlelength = 1.5)

plt.plot(t,f,'g');
```

##### Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.

```python
plt.axis([0.5, 9.5, -2.5, 2.5]);
plt.plot(t,f,'#FF5722');
```

##### Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов.

##### В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4. 
##### Дайте название графикам: 'График y1', 'График y2' и т.д.

```python
#Сделаем шришт меньше на осях
import matplotlib.pylab as pylab
params = {#'legend.fontsize': 'x-small',
          #'figure.figsize': (5, 5),
         #'axes.labelsize': 'x-small',
         #'axes.titlesize':'x-small',
         'xtick.labelsize':6,
         'ytick.labelsize':6}
pylab.rcParams.update(params)


# Чтобы было веселее сделаем цвета рандомными
import random as r
def random_color():
    return ["#"+''.join([r.choice('0123456789ABCDEF') for j in range(6)]) for i in range(4)];

ran_col = random_color();

fig,ax = plt.subplots(nrows = 2, ncols =2)
ax1, ax2, ax3, ax4 = ax.flatten()

t_dict1 = {'fontsize':8, 'color':ran_col[0]};
t_dict2 = {'fontsize':8, 'color':ran_col[1]};
t_dict3 = {'fontsize':8, 'color':ran_col[2]};
t_dict4 = {'fontsize':8, 'color':ran_col[3]};

ax1.set_title('График y1',fontdict = t_dict1, loc = 'center');
ax2.set_title('График y2',fontdict = t_dict2, loc = 'center');
ax3.set_title('График y3',fontdict = t_dict3, loc = 'center');
ax4.set_title('График y4',fontdict = t_dict4, loc = 'center');

# Уберем вернию и правую границы на графике
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax3.spines['right'].set_visible(False)
ax3.spines['top'].set_visible(False)
ax4.spines['right'].set_visible(False)
ax4.spines['top'].set_visible(False)

plt.subplots_adjust(hspace=0.5)

ax1.plot(x,y1, color = ran_col[0]);
ax2.plot(x,y2, color = ran_col[1]);
ax3.plot(x,y3, color = ran_col[2]);
ax4.plot(x,y4, color = ran_col[3]);
```

###### Объединим все условия и для лучшего восприятия поместим 2 гистограммы рядом

```python
fig, ax = plt.subplots(figsize=(8,4))
plt.hist(
    [np.clip(df1, df1[0], df1[-1]),
     np.clip(df2, df2[0], df2[-1])],
    color = ['#F80000', '#B5B8B1'],
    label = ['Class = 1', 'Class = 0'],
    log = True,
    bins = 20,
    density = True
);
plt.legend(loc = 'upper right');
plt.xlabel('Class')
fig.tight_layout();
```
